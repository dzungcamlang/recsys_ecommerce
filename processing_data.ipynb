```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./data/data.csv", index_col="Mã KH", encoding="utf-8")
```

```python
df
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df
```

```python
df = df[pd.notnull(df['Mã Sản Phẩm'])]
df = df.reset_index()
```

```python
df
```

```python
df.info()
```

```python
df["Số Lượng"]=df["Số Lượng"].apply(lambda x: x.replace(",","") if isinstance(x,str) else float(x)).astype(float)
df["Thành Tiền"]=df["Thành Tiền"].apply(lambda x: x.replace(",","") if isinstance(x,str) else float(x)).astype(float)
```

```python
df
```

```python
df.info()
```

```python
for col in df.columns:
    df[col].interpolate(method="pad", axis=0, inplace=True)
```

```python
df.sort_values(by=["Ngày Ghi Sổ"],ascending=False, inplace=True)
```

```python
df
```

```python
eng_cols = ["user_id", "order_id", "time", "description", "product_id", "product_name", "amount", "price"]
df.columns = eng_cols
df.head(2)
```

```python
df.time = pd.to_datetime(df.time, format='%m/%d/%y 0:00')
```

```python
df.info()
```

```python
df.to_csv("./data/processed_data.csv", encoding="utf-8", index=False)
```
